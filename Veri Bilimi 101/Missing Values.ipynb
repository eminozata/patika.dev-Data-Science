# Missing Values

```python
d = {"a": [1,2,3,4,4,np.nan], "b": [1,2,3,4,5,5]}
```

```python
toy = pd.DataFrame(d)
```

```python
toy
```

```python
toy["a"].isnull()
```

```python
def fix_missing(df,col,name):
    if is_numeric_dtype(col):
        if pd.isnull(col).sum:
            df[name + "_na"] = pd.isnull(col)
        df[name] = col.fillna(col.median())
```

```python
fix_missing(toy,toy["a"],"a")
```
